```python
import tensorflow.keras 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Dropout, Activation,Input,Flatten
from keras.preprocessing.text import Tokenizer
import tensorflow as tf
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.probability import FreqDist
import pandas as pd
import time
import numpy as np
import cv2
```

```python
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
dir=os.getcwd()
```

```python
train_dir=dir+"\\DevanagariHandwrittenCharacterDataset\\Train"
test_dir=dir+"\\DevanagariHandwrittenCharacterDataset\\Test"
```

```python
X_train,y_train,X_test,y_test=[],[],[],[]
```

```python
os.listdir(test_dir) 
```

```python
labels=['digit_0','digit_1','digit_2','digit_3','digit_4','digit_5','digit_6','digit_7','digit_8','digit_9']
```

```python
for i in labels:
    find=train_dir+"\\"+i
    for j in os.listdir(find) :
        img=cv2.imread(find+"//"+j)
#         print(img)
        X_train.append(img.flatten())
        y_train.append([i[-1]])
```

```python
for i in labels:
    find=test_dir+"\\"+i
    for j in os.listdir(find) :
        img=cv2.imread(find+"//"+j)
#         print(img)
        X_test.append(img.flatten())
        y_test.append([i[-1]])    
```

```python
X_train=pd.DataFrame(X_train)
```

```python
X_train.shape
```

```python
X_test=pd.DataFrame(X_test)
X_test.shape
```

```python
y_train=pd.DataFrame(y_train)
y_test=pd.DataFrame(y_test)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
```

```python
model = Sequential()
model.add(Dense(32, input_dim=3072, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=X_train.shape[0]//32,epochs=100)
```

```python
b=model.evaluate(X_train,y_train)
a=model.evaluate(X_test,y_test)
    
print(b[1])
print(a[1])
```

```python
nodes=[8, 16, 32, 64,  128,  256,  512,1024]
train_acc=[]
test_acc=[]
time_taken=[]
```

```python
for i in nodes:

    model = Sequential()
    model.add(Dense(i, input_dim=3072, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

    start=time.time()
    model.fit(X_train,y_train,batch_size=X_train.shape[0]//32,epochs=100)
    time_taken.append(time.time()-start)
    
    b=model.evaluate(X_train,y_train)
    a=model.evaluate(X_test,y_test)
    
    train_acc.append(b[1])
    test_acc.append(a[1])
```

```python
for i in range(1,5):
    
    model = Sequential()
    model.add(Dense(32, input_dim=3072, activation='relu'))
    
    for j in range(i):
        model.add(Dense(32,activation='relu'))

    model.add((Dense(1,activation='softmax')))

    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

    start=time.time()
    model.fit(X_train,y_train,batch_size=X_train.shape[0]//32,epochs=100)
    time_taken.append(time.time()-start)
    
    b=model.evaluate(X_train,y_train)
    a=model.evaluate(X_test,y_test)
    
    train_acc.append(b[1])
    test_acc.append(a[1])
```
